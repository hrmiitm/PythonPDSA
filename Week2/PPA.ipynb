## Binary Search
Write function ```BinarySearch(L,k)```  
Input: L = List of ascending order numbers, k=Element to search  
Output: (Boolean, integer) = foundOrNot, no. of comparison  

Note: middle index = (left+right)//2

````python
def BinarySearch(l, k):
    s = 0
    e = len(l)-1
    m = (s+e)//2
    (ans, c) = (False, 0)
    while(s<=e):
        c += 1
        if l[m] == k: 
            ans = True
            break
        if l[m]>k: e = m-1
        if l[m]<k: s = m+1
        m = (s+e)//2
    return (ans, c)
````

## RangeSorting

Write Function sortInRange(L, r)  
Input: L = list of intgers, r = range [0,r], r<1000, and len(L)>>>>>r  
Output: Sorted list in ascending order within TC: O(N+R)  

````python
def sortInRange(l, r):
    tempList = [0]*(r+1)
    for i in l:
        tempList[i] += 1
    
    index = 0
    for key in range(r):
        for i in range(tempList[key]):
            l[index] = key
            index += 1
    print(l)
````
